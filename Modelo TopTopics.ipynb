```python
import pandas as pd
from gensim import models
from gensim import corpora
```

```python
#Lectura de datos limpios
df = pd.read_csv("/content/drive/MyDrive/Proyecto-Tesis/Data/data-100000.csv")
```

```python
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize #Libreria para la tokenizacion de las palabras
from nltk.corpus import stopwords #Librerias para ls stopwords o palabras sin significado

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def process_text(text):
    # eliminamos caracteres especiales
    text = re.sub('[^A-Za-z]', ' ', text.lower())

    tokenized_text = word_tokenize(text)

    #Removemos los stopwords y reducimos las palabras a su raiz
    clean_text = [
        stemmer.stem(word) for word in tokenized_text
        if word not in stopwords.words('english')
    ]

    return clean_text
```

```python
tex = df['reviewText'].to_numpy()
```

```python
from multiprocessing import Pool

with Pool() as pool:
    processed_comments = pool.map(process_text, tex)
```

```python
# texts = [process_text(text) for text in tex]
```

```python
# texts = [process_text(text) for text in tex]
dic = corpora.Dictionary(processed_comments)
corpus = [dic.doc2bow(text) for text in processed_comments]
```

```python
model = models.ldamodel.LdaModel(corpus, num_topics=20, id2word=dic, passes=15)

topics = model.print_topics(num_words=3)
for topic in topics:
    print(topic)
```

```python
# Guardar el modelo
model.save("modelo_lda_guardado")

# Cargar el modelo
modelo_lda_cargado = models.LdaModel.load("modelo_lda_guardado")
```

```python
topics = modelo_lda_cargado.print_topics(num_words=3)
for topic in topics:
    print(topic)
```
