
#Modelo 1: XGBoost

##Librerias


```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

##Datos


```python
#Lectura de datos limpios
df = pd.read_csv("/content/drive/MyDrive/Proyecto-Tesis/Data/data-limmpia-100000.csv")
```

```python
df.head()
```

```python
# sparamos los rating con 0 menores iguales a 3 y con 1 mayores a 3
df['rating'].value_counts()
```

```python
df.loc[df['rating'] < 3, 'rating'] = 0
df.loc[df['rating'] == 3, 'rating'] = 1
df.loc[df['rating'] > 3, 'rating'] = 2
```

```python
df['rating'].value_counts()
```

```python
df[df['rating'].isnull()]
```

```python
df['rating'] = df['rating'].astype('int64')
```

```python
df.dtypes
```

# Modelo XGBoost

```python
# Separamos la data en train y test
X, y = df.drop("rating", axis=1), df[['rating']]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=1, stratify=y)
```

```python
xgb_classifier = xgb.XGBClassifier(n_estimators=100, objective='multi:softmax',num_class = 3, tree_method='hist', eta=0.1, max_depth=100)
hist = xgb_classifier.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)
```

```python
xgb_classifier.score(X_test, y_test)
```
