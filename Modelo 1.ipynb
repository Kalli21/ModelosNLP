# Modelo XGBoost

```python
# Separamos la data en train y test
X, y = df.drop("rating", axis=1), df[['rating']]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
xgb_classifier = xgb.XGBClassifier(n_estimators=100, objective='multi:softmax',num_class = 3, tree_method='hist', eta=0.1, max_depth=100)
hist = xgb_classifier.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)
```

```python
xgb_classifier.score(X_test, y_test)
```
